```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from matplotlib import patches
# import matplotlib.pyplot as plt
from tqdm import tqdm


def draw_rect_bbox(bbox, ax, color):
    (x0, y0, x1, y1) = bbox
    ax.add_patch(patches.Rectangle((x0, y0), x1 - x0, y1 - y0, fill=False, color=color))
```

```python
rects = [(1,1,4,4),(5,5,8,9)]
size = 6
# fig, ax = plt.subplots(figsize=(size, size * (float(y_max) / float(x_max))))
fig, ax = plt.subplots(figsize=(size*8, size * (9)))
# fig, ax = plt.subplots()
for l in tqdm(rects):
    # l_bbox = (float(l["x0"]), float(l["y0"]), float(l["x1"]), float(l["y1"]))
    l_bbox = (float(l[0]), float(l[1]), float(l[2]), float(l[3]))
    print(l_bbox)
    draw_rect_bbox(l_bbox, ax, "red")
```

```python
### 三维图
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(10, 5))  
ax = Axes3D(fig)

X3 = np.linspace(-10,10,100)
Y3 = np.linspace(-10,10,100)
X3, Y3 = np.meshgrid(X3, Y3)
Z3 = X3*X3 + Y3*Y3 + X3 + Y3 + 1
ax.plot_surface(X3, Y3, Z3, cmap=plt.cm.winter)

plt.show()
```
